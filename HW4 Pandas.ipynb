## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]
AgeAnalysis = pd.DataFrame(purchase_data[["Age"]])
AgeCount = int(AgeAnalysis.count())
AgeAnalysis["Age Category"] = pd.cut(AgeAnalysis["Age"], bins, labels= group, include_lowest = True)
AgeGroup = AgeAnalysis.groupby(["Age Category"])
AgeSummary = pd.DataFrame(AgeGroup.count())
AgeSummary["Percentage of Players"] = (AgeSummary["Age"]/AgeCount*100).map("{:.2f}%".format)
AgeSummary = AgeSummary.rename(columns={"Age":"Total Count"})
AgeSummary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
AgeAnalysis["SN"] = purchase_data["SN"]
AgeSummary2 = AgeSummary
del AgeSummary2["Percentage of Players"]
#AgeSummary2
```

```python
AgeAnalysis["Price"] = purchase_data["Price"]
test = AgeAnalysis.groupby(["Age Category"])
AvgAgePrice = test["Price"].mean()
TtlAgeValue = test["Price"].sum()
SNcount = test["SN"].nunique()

AvgAgePerson = TtlAgeValue / SNcount
AgeSummary2["Average Purchase Price"] = AvgAgePrice.map("${:.2f}".format)
AgeSummary2["Total Purchase Value"] = TtlAgeValue.map("${:.2f}".format)
AgeSummary2["Avg Total Purchase per Person"] = AvgAgePerson.map("${:.2f}".format)
AgeSummary2

#AvgAgePrice
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender = purchase_data[["SN","Price"]]
spenderGroup = spender.groupby(["SN"], sort = False)
spenderCount = spenderGroup.count()
spenderAvgPrice = spenderGroup["Price"].mean()
spenderTtl = spenderGroup["Price"].sum()#.sort_values(ascending = False)

spenderCount["Average Purchase Price"] = spenderAvgPrice.map("${:.2f}".format)
spenderCount["Total Purchase Value"] = spenderTtl.map("${:.2f}".format)
spenderCount = spenderCount.rename(columns = {"Price" : "Purchase count"})


spenderCount2 = spenderCount.sort_values("Total Purchase Value", ascending = False)


spenderCount2


```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular = purchase_data[["Item ID", "Item Name", "Price"]]
popularGroup = popular.groupby(["Item ID", "Item Name"])
popularSummary = pd.DataFrame(popularGroup.count())
price = popularGroup["Price"].mean()
popularSummary["Item Price"] = price#.map("${:.2f}".format)
popularTotal = (popularSummary["Item Price"] * popularSummary["Price"])#.map("${:.2f}".format)
popularSummary["Total Purchase Value"] = popularTotal
popularSummary2 = popularSummary.rename(columns = {"Price": "Purchase Count"})
popularSummary3 = popularSummary2.sort_values("Purchase Count", ascending = False)
finalSummary = pd.DataFrame(popularSummary3.loc[:,"Purchase Count"])
finalSummary["Item Price"]= popularSummary3["Item Price"].map("${:.2f}".format)
finalSummary["Total Purchase Value"] = popularSummary3["Total Purchase Value"].map("${:.2f}".format)
finalSummary



```
